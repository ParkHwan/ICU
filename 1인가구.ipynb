```python
householdes.columns=['2016','2017','2018','2019','2020']
householdes=householdes.drop(0)
householdes
```

```python
householdes=householdes.iloc[:3]
householdestp=householdes.transpose()
householdestp.columns=['총','남자','여자']
householdestp
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 그림 사이즈, 바 굵기 조정
fig, ax = plt.subplots(figsize=(12,8))
bar_width = 0.3

# 연도가 4개이므로 0, 1, 2, 3 위치를 기준으로 삼음
index = np.arange(5)

# 각 연도별로 3개 샵의 bar를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
b1 = plt.bar(index, householdestp['남자'], bar_width, alpha=0.4, color='red', label='man')

b2 = plt.bar(index + bar_width, householdestp['여자'], bar_width, alpha=0.4, color='blue', label='woman')

b3 = plt.bar(index + 2 * bar_width, householdestp['총'], bar_width, alpha=0.4, color='green', label='total')

# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 year 정보와 매칭

plt.xticks(np.arange(bar_width, 5 + bar_width), householdestp.index)



# x축, y축 이름 및 범례 설정
plt.xlabel('year')
plt.ylabel('num')

plt.legend()
plt.show()
```
