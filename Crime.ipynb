```python
import pandas as pd
crime = pd.read_csv('data/crime.csv')
crime.columns=["죄종별","항목","단위", "2016 년",	"2017 년",	"2018 년",	"2019 년",	"2020 년"]
crime=crime.drop(0)
crime.head(3)
```

```python
total_crime=crime[crime['죄종별'] == "총계"]
total_crime
```

```python
total_crime = total_crime[total_crime['항목'] == '발생건수[건]' ]
total_crime=total_crime.iloc[:,3:]
total_crime=total_crime.transpose()
total_crime[1] = pd.to_numeric(total_crime[1])
total_crime
```

```python
import matplotlib.pyplot as plt
import numpy as np
## 데이터

x = np.arange(5)
years  = total_crime.index 
values = total_crime[1] 
 
plt.bar(x, values)
plt.xticks(x, years)
plt.title("Total_crime")

plt.show()
```
