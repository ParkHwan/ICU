```python
import numpy as np
import pandas as pd
import folium
```

```python
crime=pd.read_csv('data/crime(5.csv')
crime=crime.iloc[:,1:3]
crime=crime.iloc[2:]
crime["합계"] = pd.to_numeric(crime["합계"].str.replace(',', ''))
crime
```

```python
import json
seoul_geo = json.load(open('data/geodata.json', encoding='utf-8'))
seoul_map = folium.Map(location=[37.5502, 126.982],zoom_start=11,tiles='Stamen Toner')

folium.Choropleth(
    geo_data=seoul_geo,
    data=crime,
    columns=['자치구', '합계'],
    key_on='feature.properties.name',
    fill_color='YlGn', 
    legend_name='자치구별 5대 범죄'
).add_to(seoul_map)

seoul_map
```

```python
pop_seoul = pd.read_csv("data/인구__가구_및_주택__읍면동_2015_2020___시군구_20162019__20220603165301.csv", encoding="cp949")
pop_seoul_2020 = pd.DataFrame(pop_seoul[["행정구역별(읍면동)", "2020", "2020.1", "2020.2"]])
column1 = ["행정구역별", "총인구(명)", "남자(명)", "여자(명)"]
pop_seoul_2020.drop(0, inplace = True)
pop_seoul_2020.rename(columns={'행정구역별(읍면동)':'자치구'},inplace=True)
pop_seoul_2020.head(2)
```

```python
pop_crime_seoul = pd.merge(pop_seoul_2020, crime, how="left", on="자치구")
pop_crime_seoul=pop_crime_seoul.iloc[1:,]
pop_crime_seoul[["2020", "2020.1", "2020.2","합계"]] = pop_crime_seoul[[ "2020", "2020.1", "2020.2","합계"]].astype(int)
pop_crime_seoul.head(2)
```

```python
pop_crime_seoul["인구10만명당범죄"] = pop_crime_seoul["합계"] / pop_crime_seoul["2020"] * 100000
pop_crime_seoul.head(2)
```

```python
import json
seoul_geo = json.load(open('data/geodata.json', encoding='utf-8'))
seoul_map = folium.Map(location=[37.5502, 126.982],zoom_start=11,tiles='Stamen Toner')

folium.Choropleth(
    geo_data=seoul_geo,
    data=pop_crime_seoul,
    columns=['자치구', '인구10만명당범죄'],
    key_on='feature.properties.name',
    fill_color='YlGn', 
    legend_name='인구 10만명당 5대 범죄발생건수'
).add_to(seoul_map)

seoul_map
```
